## K-Means Example: Tuesday March 19th, 2019

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('tips')

df.head()
```

```python
df.shape
```

```python
from sklearn.cluster import KMeans

X = df[['total_bill', 'tip', 'size']]

model= KMeans(2)

model.fit(X)

cluster_labels = model.predict(X)
```

```python
cluster_labels
```

```python
# an also combine into one step

from sklearn.cluster import KMeans

X = df[['total_bill', 'tip', 'size']]

model= KMeans(2)

cluster_labels = model.fit_predict(X)
```

```python
cluster_labels
```

```python
df['clusters'] = cluster_labels
```

```python
df
```

```python
df.groupby('clusters')['tip'].mean()
```

## Elbow method

```python
from scikitplot.cluster import plot_elbow_curve

model = KMeans()

plot_elbow_curve(model, X);
```

```python
model = KMeans(5, random_state=1)

cluster_labels = model.fit_predict(X)

df['clusters'] = cluster_labels
```

```python
df
```

```python
df.groupby('clusters')['tip'].mean()
```

## Silhouette method

```python
from scikitplot.metrics import plot_silhouette
```

```python
plot_silhouette(X, cluster_labels);
```

```python
model = KMeans(4, random_state=1)

cluster_labels = model.fit_predict(X)

df['clusters'] = cluster_labels

plot_silhouette(X, cluster_labels);
```

```python
from sklearn.cluster import KMeans

X = df[['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']]

model= KMeans(2)

cluster_labels = model.fit_predict(X)
```
